```python
# create Tensor1, 2, 3

t1 = torch.FloatTensor([[1,2,3,4], [10,20,30,40]])
t2 = torch.FloatTensor([[100,200,300,400], [50,60,70,80]])
t3 = torch.FloatTensor([[31,32,33,34],[20,21,22,23]])
```

```python
print("====Concatenate of tensors====")

print(f"default cat : \n{torch.cat([t1,t2,t3])}\n")
print(f"dim 0 cat : \n{torch.cat([t1,t2,t3], dim=0)}\n")
print(f"dim 1 cat : \n{torch.cat([t1,t2,t3], dim=1)}\n")
```

```python
print("====Stack of tensors====")

print(f"default stack : \n{torch.stack([t1,t2,t3])}\n")
print(f"dim 0 stack : \n{torch.stack([t1,t2,t3], dim=0)}\n")
print(f"dim 1 stack : \n{torch.stack([t1,t2,t3], dim=1)}\n")
print(f"dim 2 stack : \n{torch.stack([t1,t2,t3], dim=2)}\n")
```
