# to_datetime
- 시간 관련 데이터를 Python의 datetime 객체로 변환
- 다양한 형태의 날짜 및 시간 데이터를 처리할 수 있음
```python
pd.to_datetime(df['col'])
pd.datetime.strptime(df['col'], "%Y%m%d%H%M")  # datetime 포맷에서 원하는 형태 str으로 변환
```

````python
import pandas as pd

df = pd.read_csv('taxis.csv')
df.head(2)
df.info()
````

````python
df['pickup'] = pd.to_datetime(df.pickup)
df.head(2)
````

````python
df.info()
````

# to_period
- 날짜/시간 데이터를 주어진 주기('M', 'W', 'Y' 등)로 변환
```python
 df['col'].dt.to_period('period').dt.strftime('%Y-%m-%d')
# to_period('period') : Daily(D), Weekly(W), Monthly(M), Yearly(Y) 등 주기로 나타냄
# dt.strftime('%Y-%m-%d') : 'YYYY-MM-DD' 형태로 표시함
 ```
 [python string format doc](https://docs.python.org/3/library/datetime.html#strftime-and-strptime-behavior)

````python
print(df.pickup.dt.to_period('W')[0])
# df.dropoff.dt.to_period 하면 오류남. datetime dtype이 아니라서.

# 주의 마지막 날짜로 표시됨
df['pickup_w'] = df.pickup.dt.to_period('W').dt.strftime("%Y-%m-%d")

# 주의 시작 날짜로 표시됨
df['pickup_w'] = df.pickup.dt.to_period('W').dt.start_time

df.head(2)
````

````python
df.info()
````

````python
import pandas as pd
from datetime import datetime

dt_parser = lambda x: datetime.strptime(x, '%Y-%m-%d %H:%M:%S')

df = pd.read_csv('taxis.csv',
                 parse_dates=['pickup', 'dropoff'],
                 date_parser=dt_parser)

df.info()
````

1. for 루프를 사용하여 DataFrame의 'start' 열과 'end' 열을 datetime 유형으로 변환하십시오.
2. start에서 end까지의 시간 차이를 시간 단위로 계산하십시오. 시간 이하 단위의 차이는 소수점으로 표시하십시오.

````python
# 1. for 루프를 사용하여 DataFrame의 'start' 열과 'end' 열을 datetime 유형으로 변환하십시오.

import pandas as pd

data = {
    'start': ['2021-01-01 10:00', '2021-01-02 11:30', '2021-01-03 09:45', '2021-01-04 15:00', '2021-01-05 18:20'],
    'end': ['2021-01-01 15:30', '2021-01-02 01:00', '2021-01-04 10:15', '2021-01-05 15:30', '2021-08-05 18:50']
}
# 데이터프레임 생성
df = pd.DataFrame(data)
df.info()
````

````python

cols = ['start', 'end']
for col in cols:
    df[col] = pd.to_datetime(df[col])

df.info()
````

````python
# 2. start에서 end까지의 시간 차이를 시간 단위로 계산하십시오. 시간 이하 단위의 차이는 소수점으로 표시하십시오.

(df.end - df.start).dt.total_seconds()/3600
````

````python
# 3. 'end' 열의 날짜들을 주 단위('W')로 변환하고, 이를 'yyyy-mm-dd' 형태의 문자열로 표현하십시오.

# df.end.dt.to_period('W').dt.strftime('%Y-%m-%d')
df.end.dt.to_period('W').dt.start_time
````
