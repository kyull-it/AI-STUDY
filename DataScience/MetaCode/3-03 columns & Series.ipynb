  - CustomerID: Í≥†Í∞ùÏùò Í≥†Ïú† ID

  - Product: Íµ¨Îß§Ìïú ÏÉÅÌíàÎ™Ö

  - Ïª¨Îüº ÏÇ≠Ï†ú: OrderDate Ïª¨ÎüºÏùÑ ÏÇ≠Ï†úÌïòÏÑ∏Ïöî.
  - ÏÉà Ïª¨Îüº Ï∂îÍ∞Ä: Í∞Å Ï£ºÎ¨∏Ïùò Ï¥ù Í∏àÏï°ÏùÑ ÎÇòÌÉÄÎÇ¥Îäî TotalAmountÎùºÎäî ÏÉà Ïª¨ÎüºÏùÑ Ï∂îÍ∞ÄÌïòÏÑ∏Ïöî. Ïù¥ Ïª¨ÎüºÏùÄ QuantityÏôÄ PriceÏùò Í≥±ÏúºÎ°ú Í≥ÑÏÇ∞Îê©ÎãàÎã§.

````python
import pandas as pd

# ÏòàÏ†ú Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ ÏÉùÏÑ±
data = {
    'CustomerID': [101, 102, 103, 104],
    'OrderDate': ['2021-01-01', '2021-01-02', '2021-01-03', '2021-01-04'],
    'Product': ['Apple', 'Banana', 'Cherry', 'Date'],
    'Quantity': [5, 3, 4, 2],
    'Price': [100,200,250,50]
}

df = pd.DataFrame(data)
df
````

````python
df.rename(columns={'CustomerID':'ID'}, inplace=True)
df.drop('OrderDate', axis=1, inplace=True)
df['TotalAmount'] = df['Quantity'] * df['Price']
df
````

````python
import pandas as pd

df = pd.read_csv('./iris.csv')
df.head()
````

````python
df.SepalLengthCm
df['SepalLengthCm']
df[['SepalLengthCm', 'SepalWidthCm', 'Species']]
df.drop('Id', axis=1, inplace=True)
df.rename(columns={'SepalLengthCm':'SL', 'SepalWidthCm':'SW', 'PetalLengthCm':'PL', 'PetalWidthCm':'PW'}, inplace=True)
df['total'] = df['SL'] + df['SW'] + df['PL'] + df['PW']
df.head()
````

````python
print(f"Type of df['SL'] : {type(df['SL'])}")
print(f"Type of df['SL'].dtype : {df['SL'].dtype}")
print("min of SL\t: ", df['SL'].min())
print("max of SL\t: ", df['SL'].max())
print("sum of SL\t: ", df['SL'].sum())
print("mean of SL\t: ", df['SL'].mean())
print("unique of species\t: ", df['Species'].unique())
print("the number of species\t: ", df['Species'].nunique())
````

````python
print("value_counts of species\t: ", df['Species'].value_counts().head())
print()
print("nlargest of SL\t: ", df['SL'].nlargest(5))
print()
print(df.nlargest(5, 'SL'))
print()
print(df[['SL', 'SW']].value_counts().head())
````
