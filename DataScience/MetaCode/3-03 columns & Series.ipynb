# column 수정하기

```python
# inplace = True: 원본 df 변경
```

## column명 변경: rename()

# Series
- DataFrame 에서 하나의 column 만 선택하면 series

````python
type(df_iris['sepal_length'])     # pandas.core.series.Series

df_iris['sepal_length'].dtype     # dtype('float64') : 실수형

````

````python
# 기본 집계 min, max, sum, mean

print(df_iris['sepal_length'].min())
print(df_iris['sepal_length'].max())
print(df_iris['sepal_length'].sum())
print(df_iris['sepal_length'].mean())
````

- unique(): 중복 없는 값들 나열.
- nunique(): 중복 없는 값들 수 세기.


````python
# unique, nunique
print(df_iris['species'].unique())
print(df_iris['species'].nunique())
````

````python
# value_counts
print(df_iris['species'].value_counts())
````

# nlargest(), nsmallest()
```python
df.nlargest(number,['col1','col2',...])
```

````python
# nlargest 다중 나열
print(df_iris['sepal_length'].nlargest(5))
print()
print(df_iris.nlargest(5, 'sepal_length'))
````

# value_counts()
- value_counts(): 각 값 별로 개수 세기.
- count가 큰 순서대로 정렬됨.

````python
# column 하나
print(df_iris['sepal_length'].value_counts().head())
````

````python
# column 두개
df_iris[['sepal_length', 'sepal_width']].value_counts().head()
````

````python
# plot() bar 차트
df_iris['sepal_length'].value_counts().plot(kind='bar')
````

2. 데이터셋 설명
- 데이터셋은 가상의 온라인 상점의 고객 주문 내역을 포함하고 있다고 가정합니다. 데이터셋에는 다음과 같은 컬럼이 있습니다:

  - CustomerID: 고객의 고유 ID
  - OrderDate: 주문 날짜
  - Product: 구매한 상품명
  - Quantity: 구매한 상품의 수량
  - Price: 상품의 가격
- 예제 작업
  - 컬럼 이름 변경: CustomerID 컬럼의 이름을 ID로 변경하세요.
  - 컬럼 삭제: OrderDate 컬럼을 삭제하세요.
  - 새 컬럼 추가: 각 주문의 총 금액을 나타내는 TotalAmount라는 새 컬럼을 추가하세요. 이 컬럼은 Quantity와 Price의 곱으로 계산됩니다.

````python
import pandas as pd

# 예제 데이터프레임 생성
data = {
    'CustomerID': [101, 102, 103, 104],
    'OrderDate': ['2021-01-01', '2021-01-02', '2021-01-03', '2021-01-04'],
    'Product': ['Apple', 'Banana', 'Cherry', 'Date'],
    'Quantity': [5, 3, 4, 2],
    'Price': [100,200,250,50]
}

df = pd.DataFrame(data)
df
````

````python
df.rename(columns={'CustomerID':'ID'}, inplace=True)
df.drop('OrderDate', axis=1, inplace=True)
df['TotalAmount'] = df['Quantity'] * df['Price']
df
````
