# For Loop
- for 루프는 주로 시퀀스(리스트, 문자열 등 순회할 대상)를 순회하며, 각 원소에 대해 루프 내부의 코드를 실행
```python
for 변수 in 시퀀스:
    # 실행할 코드
```

````python
# 리스트란 : 순서가 있는 아이템들의 모음 [](대괄호)로 표현
my_list = [1,2,3,4]
````

````python
# 0에서 부터 시작( 왼쪽 ) , -1 부터 시작 ( 오른쪽 )
````

````python
seq = 0
for i in [1, 10, 100, 1000] :
  seq = seq +  1  # seq += 1 과 동일
  print('---start---')
  print(f'시쿼스: {seq}, 아이템: {i}')
  print( '---end---' )
  print('')
# https://pythontutor.com/visualize.html#mode=edit   코드 동작하는 순서
````

# While Loop
- 조건이 참(True)인 동안 반복적으로 코드 블록을 실행하는 데 사용됩니다
```python
while 조건:
    # 실행할 코드
```

````python
i = 0
j = 1
while i < 5:
  bool = i < 5
  i = i + 1
  j = j*10
  print(f'i:{i}, Boolean: {bool}, j:{j}')
print( f'boolean: {i < 5}'  )
# https://pythontutor.com/visualize.html#mode=edit   코드 동작하는 순서
````
