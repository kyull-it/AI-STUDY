# Index
- 인덱스 (Index): 행을 구별하는 데 사용되는 고유한 식별자.

```python
df.set_index('col1', inplace = True) # index 지정
df.reset_index(inplace = True ) # 기본 정수 Index
```

````python
# taxis
import seaborn as sns
df = sns.load_dataset('taxis')
df.head()
````

````python
df.set_index('pickup', inplace=True)
df.head()
````

````python
df.reset_index(inplace=True)
df.head()
````

# Sort
- 특정 기준에 따라 순서대로 `나열`
- sort_value():특정 column 기준
```python
df.sort_values('col1', ascending = False, inplace = True )
```
  - ascending = False: 내림차순
  - ascending = True: 오름차순

- sort_index(): index 기준
```python
df.sort_index(inplace= True)
```

````python
df.sort_values('distance', ascending=False).head()
````

````python
df.sort_values('distance', ascending=False).reset_index().head()
````

````python
df.sort_values('distance', ascending=False, inplace=True)
df.head()
````

````python
# 이전에 reset_index()를 호출하지 않았기 때문에, 기존의 df index가 그대로 남아있음. 
df.sort_index().head()
````

```python

import pandas as pd

# 데이터프레임 생성
data = {
    'ProductID': [101, 102, 103, 104],
    'SaleDate': ['2021-07-01', '2021-07-03', '2021-07-02', '2021-07-04'],
    'Quantity': [3, 2, 5, 4]
}

df = pd.DataFrame(data)
```

- ProductID를 데이터프레임의 인덱스로 설정하세요.
- Quantity를 기준으로 데이터를 내림차순으로 정렬하세요.
- 정렬된 데이터프레임의 인덱스를 기본 인덱스로 재설정하세요.

````python
import pandas as pd

data = {
    'ProductID': [101, 102, 103, 104],
    'SaleDate': ['2021-07-01', '2021-07-03', '2021-07-02', '2021-07-04'],
    'Quantity': [3, 2, 5, 4]
}

df = pd.DataFrame(data)
df
````

````python
df.set_index('ProductID', inplace=True)
df.sort_values('Quantity', ascending=False, inplace=True)
df
````
