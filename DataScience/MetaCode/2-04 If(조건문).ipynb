# Boolean
- 참(True)과 거짓(False)만을 나타내는 데이터 타입
  - 비교 연산자
    - == (Equal to): 두 값이 `같은지` 비교합니다. 같으면 참(True)을 반환합니다.
    - != (Not equal to): 두 값이 `다른지` 비교합니다. 다르면 참을 반환합니다.
    - \> (Greater than): 왼쪽 값이 오른쪽 값보다 `큰지` 비교합니다. 크면 참을 반환합니다.
    - \< (Less than): 왼쪽 값이 오른쪽 값보다 `작은지` 비교합니다. 작으면 참을 반환합니다.
    - \>= (Greater than or equal to): 왼쪽 값이 오른쪽 값보다 `크거나 같은지` 비교합니다. 크거나 같으면 참을 반환합니다.
    - \<= (Less than or equal to): 왼쪽 값이 오른쪽 값보다 `작거나 같은지` 비교합니다. 작거나 같으면 참을 반환합니다.

  ```python
    True # 대문자로 적줘야 합니다.
    False
    ---
    true <- x
    false <- x
  ```

````python
equal = 1 == 1
````

````python
print( equal )
````

````python
not_equal = 1 == 2
````

````python
print( not_equal)
````

# IF(조건문)
- 조건문은 프로그래밍에서 "만약 이 조건이 참이라면 이 코드를 실행하라"라는 의미를 가진 구조
```python
  if 조건:
    # 조건이 참(True)일 때 실행되는 코드
  else:
    # 조건이 거짓(False)일 때 실행되는 코드
```


````python
if True  : # 1>2 | 1<2 | True | False | true | false
  print('맞다')
else:
  print( '틀리다')
````

## elif

  ```python
  if 조건1:
      # 조건1이 참(True)일 때 실행되는 코드
  elif 조건2:
      # 조건1이 거짓(False)이고, 조건2가 참(True)일 때 실행되는 코드
  else:
      # 모든 조건이 거짓(False)일 때 실행되는 코드
  ```

````python
if False :  #조건1
  print("조건1이 맞다")
elif False : # 조건2
  print("조건1은 틀리고 조건2가 맞다")
else:
  print( "둘다 틀리다")
````
