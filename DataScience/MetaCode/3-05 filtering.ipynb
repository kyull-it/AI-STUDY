## between(), isin() , isna()
```python

```

````python
#titannic btw age
import seaborn as sns
titanic = sns.load_dataset('titanic')

con = titanic.age.between(20,22)
sorted_df = pd.DataFrame(titanic[con])
sorted_df.sort_values('age', inplace=True)
sorted_df.reset_index(drop=True, inplace=True)
print(sorted_df.value_counts('age'))
sorted_df.head()
````

````python
#isin embarked
con = titanic.embarked.isin(['C','Q'])
sorted_df = pd.DataFrame(titanic[con])
sorted_df.reset_index(drop=True, inplace=True)
print(sorted_df.value_counts('embarked'))
sorted_df.head()
````

````python
#isna deck ~ ÏÇ¨Ïö©: Î∞òÎåÄÎ°ú
print(titanic.deck.value_counts(dropna=False))

con = titanic.deck.isna()
sorted_df = pd.DataFrame(titanic[con])
print(f"Í≤∞Ï∏°Ïπò Í∞úÏàò : ", len(sorted_df))
sorted_df.head()
````

````python
# ~ ÏÇ¨Ïö© : Î∞òÎåÄÎ°ú Í≤∞Ï∏°ÏπòÍ∞Ä ÏïÑÎãå Îç∞Ïù¥ÌÑ∞Îßå Ï∂îÏ∂ú
con = titanic.deck.isna()
sorted_df = pd.DataFrame(titanic[~con])
print(f"Í≤∞Ï∏°Ïπò Í∞úÏàò : ", len(sorted_df))
sorted_df.head()
````

## and(&), or(|)
- Îëê Í∞ú Ïù¥ÏÉÅÏùò Ï°∞Í±¥ÏùÑ Í≤∞Ìï©
  - and: Î™®Îëê True -> True
  - or: ÌïòÎÇòÎùºÎèÑ True -> True


````python
# True
# True or False
# True | False
# True and True
# True & True

# False
# True and False
# False or False
````

````python
# ex age, deck
titanic['col1'] = titanic.age > 30
titanic['col2'] = titanic.deck.isna()
titanic['col3'] = titanic.age > 30 & titanic.deck.isna()
titanic['col4'] = (titanic.age > 30) & (titanic.deck.isna())
titanic['col5'] = (titanic.age > 30) | (titanic.deck.isna())
titanic['col6'] = ~(titanic.age > 30)
titanic['col7'] = ~(titanic.deck.isna())

titanic.head()
````

````python
import seaborn as sns
titanic = sns.load_dataset('titanic')
titanic.head()
````

````python
#   - classÍ∞Ä 'First'Ïù¥Í≥†, ageÍ∞Ä 30 Ïù¥ÏÉÅÏù∏ ÏäπÍ∞ùÎì§Ïùò Îç∞Ïù¥ÌÑ∞Î•º ÌïÑÌÑ∞ÎßÅÌïòÏÑ∏Ïöî.

# con = (titanic['class'].isin(['First'])) & (titanic.age > 30)
filter1 = (titanic['class'] == 'First') & (titanic.age > 30)
sorted_df = pd.DataFrame(titanic[filter1])
sorted_df.head()
````

````python
# - survivedÍ∞Ä 1(ÏÉùÏ°¥)Ïù¥Í±∞ÎÇò embark_townÏù¥ 'Southampton'Ïù∏ ÏäπÍ∞ùÎì§Ïùò Îç∞Ïù¥ÌÑ∞Î•º ÌïÑÌÑ∞ÎßÅÌïòÏÑ∏Ïöî.
filter2 = (titanic['survived'] == 1) | (titanic.embark_town == 'Southampton')
sorted_df = pd.DataFrame(titanic[filter2])
sorted_df.head()
````

````python
# - fareÍ∞Ä 100 Ïù¥ÏÉÅÏù¥Í≥†, sexÍ∞Ä 'female'Ïù∏ ÏäπÍ∞ùÎì§Ïùò Îç∞Ïù¥ÌÑ∞Î•º ÌïÑÌÑ∞ÎßÅÌïòÏÑ∏Ïöî.

filter3 = (titanic['fare'] >= 100) & (titanic['sex'] == 'female')
sorted_df = pd.DataFrame(titanic[filter3])
sorted_df.head()
````
