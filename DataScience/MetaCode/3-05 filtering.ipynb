# Filtering
- pd.DataFrame 에서 Filtering 은 조건에 부합(Boolean 값이 True)하는 row 만 남기는 과정
```python
df[조건]
# boolean-series:[True, False, ..., True]
```
  


````python
# 예시
import pandas as pd

# 예제 데이터프레임 생성
data = {
    'Name': ['철수', '영희', '흥민', '범근'],
    'Age': [25, 30, 35, 40],
    'City': ['New York', 'Los Angeles', 'New York', 'Chicago']
}

df = pd.DataFrame(data)
````

````python
# df.Age >  30; age_over_30
con = df.Age > 30
print(con)
df.Age > 30
````

````python
# filter result
# age_over_30 = df[df.Age > 30]
age_over_30 = df[con]
age_over_30
````

## between(), isin() , isna()
```python
#from_n 이상, to_n 이하
df['col'].between(from_n,to_n)

# 특정 열(col)의 값이 주어진 목록(['item1', 'item2'])에 포함되어 있는지 여부
df['col'].isin(['item','item'])

# 특정 열(col)의 값이 결측치(NA)일 경우 True, 아닐 경우 False를 반환
df['col'].isna()


```

````python
#titannic btw age
import seaborn as sns
titanic = sns.load_dataset('titanic')

con = titanic.age.between(20,22)
sorted_df = pd.DataFrame(titanic[con])
sorted_df.sort_values('age', inplace=True)
sorted_df.reset_index(drop=True, inplace=True)
print(sorted_df.value_counts('age'))
sorted_df.head()
````

````python
#isin embarked
con = titanic.embarked.isin(['C','Q'])
sorted_df = pd.DataFrame(titanic[con])
sorted_df.reset_index(drop=True, inplace=True)
print(sorted_df.value_counts('embarked'))
sorted_df.head()
````

````python
#isna deck ~ 사용: 반대로
print(titanic.deck.value_counts(dropna=False))

con = titanic.deck.isna()
sorted_df = pd.DataFrame(titanic[con])
print(f"결측치 개수 : ", len(sorted_df))
sorted_df.head()
````

````python
# ~ 사용 : 반대로 결측치가 아닌 데이터만 추출
con = titanic.deck.isna()
sorted_df = pd.DataFrame(titanic[~con])
print(f"결측치 개수 : ", len(sorted_df))
sorted_df.head()
````

## and(&), or(|)
- 두 개 이상의 조건을 결합
  - and: 모두 True -> True
  - or: 하나라도 True -> True


````python
# True
# True or False
# True | False
# True and True
# True & True

# False
# True and False
# False or False
````

````python
# ex age, deck
titanic['col1'] = titanic.age > 30
titanic['col2'] = titanic.deck.isna()
titanic['col3'] = titanic.age > 30 & titanic.deck.isna()
titanic['col4'] = (titanic.age > 30) & (titanic.deck.isna())
titanic['col5'] = (titanic.age > 30) | (titanic.deck.isna())
titanic['col6'] = ~(titanic.age > 30)
titanic['col7'] = ~(titanic.deck.isna())

titanic.head()
````

1. 연습문제: 타이타닉 데이터셋 필터링
- 문제:아래는 타이타닉 호 승객 데이터에 대한 간략한 설명과 함께 필터링 작업을 수행하는 문제입니다. Seaborn 라이브러리를 사용하여 타이타닉 데이터셋을 불러온 후, 다음 작업을 수행하세요.
  - class가 'First'이고, age가 30 이상인 승객들의 데이터를 필터링하세요.
  - survived가 1(생존)이거나 embark_town이 'Southampton'인 승객들의 데이터를 필터링하세요.
  - fare가 100 이상이고, sex가 'female'인 승객들의 데이터를 필터링하세요.



````python
import seaborn as sns
titanic = sns.load_dataset('titanic')
titanic.head()
````

````python
#   - class가 'First'이고, age가 30 이상인 승객들의 데이터를 필터링하세요.

# con = (titanic['class'].isin(['First'])) & (titanic.age > 30)
filter1 = (titanic['class'] == 'First') & (titanic.age > 30)
sorted_df = pd.DataFrame(titanic[filter1])
sorted_df.head()
````

````python
# - survived가 1(생존)이거나 embark_town이 'Southampton'인 승객들의 데이터를 필터링하세요.
filter2 = (titanic['survived'] == 1) | (titanic.embark_town == 'Southampton')
sorted_df = pd.DataFrame(titanic[filter2])
sorted_df.head()
````

````python
# - fare가 100 이상이고, sex가 'female'인 승객들의 데이터를 필터링하세요.

filter3 = (titanic['fare'] >= 100) & (titanic['sex'] == 'female')
sorted_df = pd.DataFrame(titanic[filter3])
sorted_df.head()
````
