- 주단위로 유입경로별 성과를 트레킹

### CPC : Cost Per Click
### ROAS (Return On Advertising Spend)

- **정의**: 광고 지출에 대한 수익을 나타내는 마케팅 지표.

- **계산 방법**:
$$
\text{ROAS} = \frac{\text{광고로 인한 수익}}{\text{광고 지출}}
$$

- **중요성**:
  - **높은 ROAS**: 광고 지출 대비 높은 수익을 의미. 광고 캠페인의 효과성을 나타냄.
  - **낮은 ROAS**: 광고 투자에 대한 낮은 수익을 의미. 광고 전략의 개선이 필요함을 시사.
