# DataFrame 조사


- head(): 처음 5줄 보여줌.
- tail(): 마지막 5줄 보여줌.
- info(): 열 정보, 데이터 타입 요약.
- describe(): 숫자형 열 통계 요약.

````python
# seaborn 으로 데이티 가져오기
# https://github.com/mwaskom/seaborn-data
import seaborn as sns
df = sns.load_dataset('iris')
df.head()
````

````python
df.tail()
````

````python
df.info()
````

````python
df.describe()
````
