## Case Sensitivity
- Python 은 대소문자를 구분함

```python
# Python은 대소문자를 구분합니다.
variable = 10
Variable = 20

# 이 두 변수는 서로 다릅니다.
print(variable)  # 출력: 10
print(Variable)  # 출력: 20
```

# Operation(연산)
- +
- -
- *
- /: `a/b -> a 나누기 b`  ex. 5/2 = 2.5
- //: `a/b -> a 나누기 b`(소수점 아래를 버리고 정수만 반환) ex. 5/2 = 2
  
- %: `a%b -> a를 b 로 나눈 나머지`  
  

```python
# 더하기 연산 (+)
addition = 5 + 3
```

```python
print( addition )
```

```python
# 빼기 연산 (-)
subtraction = 5 - 3
```

```python
print( subtraction)
```

```python
# 곱하기 연산 (*)
multiplication = 5 * 3
```

```python
print( multiplication)
```

```python
# 나누기 연산 (/)
division = 5 / 3
```

```python
print( division)
```

```python
# 나누기 연산 정수 (//)
division2 = 5 // 3
```

```python
print( division2)
```

```python
# 나머지 연산 (%) 5를 3으로 나눈 나머지
modulus = 5 % 3
```

```python
print( modulus)
```

---

```python
# 연산의 순서
# 괄호 없이 연산
result_without_parentheses = 10 + 5 * 2

# 괄호를 사용한 연산
result_with_parentheses = (10 + 5) * 2

print( result_without_parentheses, result_with_parentheses)
```

```python
print(str1 + " " + str2)
```

# f-strings
- 변수, 표현식, 함수 호출 등을 문자열 내에 쉽게 포함

```python
name = "Alice"
age = 24
formatted_string = f"내 이름은 {name}이고 나이는 {age}살입니다."
```

```python
print( formatted_string )
```
