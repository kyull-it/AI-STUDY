- 함수 : 특정 작업을 수행하는 코드의 집합으로, 재사용 가능
  - Paremeter(매개변수) : 함수의 입력값을 지정하기 위한 변수


$ 2x +3 = f(x)$

| x | f(x)  |
|---|-------|
| 1 | 5     |
| 2 | 7     |
| 3 | 9     |
| 4 | 11    |
| 5 | 13    |



````python
# ^함수 설정
def operation1(x):
  '''
  2x + 3 연산을 수행
  x = int
  '''
  result = (2 * x) + 3
  return result
````

````python
# 함수 설명
help(operation1)
````

````python
print( operation1(1))
````
