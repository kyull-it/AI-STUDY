## 단축어

* Shift + Enter: 현재 셀을 실행하고 다음 셀로 이동합니다.
* Ctrl/Cmd + Enter: 현재 셀을 실행하고 현재 셀에 머뭅니다.

* Ctrl/Cmd + M D: 현재 셀을 삭제합니다.
* Ctrl/Cmd + M B: 현재 셀 아래에 새 셀을 삽입합니다.
* Ctrl/Cmd + M A: 현재 셀 위에 새 셀을 삽입합니다.
* Ctrl/Cmd + M M: 현재 셀을 마크다운 셀로 변경합니다.
* Ctrl/Cmd + M Y: 현재 셀을 코드 셀로 변경합니다.
* Ctrl/Cmd + /: 코드 셀에서 현재 줄 또는 선택된 줄들을 주석 처리합니다.
  - 주석은 코드에 메모를 추가하거나 코드의 특정 부분을 실행에서 제외시키기 위해 사용됩니다

* Ctrl/Cmd + M L: 현재 셀에서 줄 번호를 표시하거나 숨깁니다.
* Ctrl/Cmd + S: 현재 셀을 선택합니다.
* Ctrl/Cmd + H: 단축어 설정
* Ctrl + Shift + P : Command Palette
